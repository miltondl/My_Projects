```python
import random


used_ids = set()

def generate_unique_id():
    unique_id = random.randint(100000, 1000000)  # Adjust the range as needed
    while unique_id in used_ids:
        unique_id = random.randint(100000, 1000000)  # Adjust the range as needed
    used_ids.add(unique_id)
    return unique_id
```

```python
import pandas as pd

# Leer CSV de ejemplo
df = pd.read_csv("golden_dataset_source.csv")

# Count annotators per label and ItemID
counts = df.groupby(['ItemID', 'Label']).size().reset_index(name='count')

# Filter labels that have at leats 2 annotators
golden_df = counts[counts['count'] >= 2].copy()

# Remove count column
golden_df = golden_df[['ItemID', 'Label']]


# Adding unique ID that will be used in the analysis
golden_df["ID"] = golden_df.apply(lambda row: generate_unique_id(), axis=1)


# # Saving golden dataset
# golden_df.to_csv("data/golden_dataset.csv", index=False)
# print("Golden dataset generado en data/golden_dataset.csv")
golden_df
```

```python
import pandas as pd
import random

# Read the golden dataset
golden_df = pd.read_csv("golden_dataset.csv")

# List of taggers or QC members to assign tasks
taggers = ["Alice", "Bob", "Carla", "David"]

# Number of items to assign per tagger
samples_per_tagger = 5

# Create an empty list to store task assignments
tasks = []


# Take a random sample of items for the task
sample = golden_df.sample(n=samples_per_tagger, replace=False)

# TaskID
task_id = random.randint(100000, 999999)


# Creating task

for _, row in sample.iterrows():
# Simulate task assignment (like Dataloop)
    for tagger in taggers:
        # Take a random sample of items for this tagger
        sample = golden_df.sample(n=samples_per_tagger, replace=False)
    
        # Generate a demo image link (simulated)
        image_link = f"https://demo-images.com/{row['ItemID']}.jpg"
        # Append task info to the tasks list
        tasks.append({
            "TaskID": task_id,
            "ItemID": row["ItemID"],
            "Label": row["Label"],
            "AssignedTo": tagger,
            "Status": "Open",  # Simulate task status
            "ImageLink": image_link
        })

# Convert tasks list to a DataFrame
tasks_df = pd.DataFrame(tasks)

# Save the simulated assigned tasks to CSV
tasks_df.to_csv("assigned_tasks_simulated.csv", index=False)

print("✅ Tasks simulated successfully!")
print(tasks_df.head())
```
